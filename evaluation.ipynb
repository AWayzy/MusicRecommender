```python
# given a list of songs, it feeds each song into the recommender individually and returns a list of recommended songs
def recommendSongs(songIDs):
    songs = ["['" + i + "']" for i in songIDs]
    return [i[2:-2] for i in rules[rules["antecedents"].isin(songs)]['consequents'].values]
```

```python
print(recommendSongs(['britneyspears_toxic', 'justintimberlake_rockyourbody']))
```

```python
# returns [correct recommendations, total recommendations, length of playlist]
def eval_row(row):
    playlist = set(row[row == 1].index)
    recs = set(recommendSongs(playlist))
    return [len(recs.intersection(playlist)), len(recs), len(playlist)]
    
```

```python
row = playlists.iloc[0]
row[row == 1].index
```

```python
eval_row(playlists.iloc[0])
```

```python
results = playlists.apply(eval_row, axis = 1, result_type = 'expand')
results.columns = ['correct_recs', 'total_recs', "total_correct"]
results
```

```python
def calculate_metrics(results, min_length):
    print(f"calulating with min length {min_length}\n")
    results = results[results['total_correct'] >= min_length]
    total = results.sum(axis=0)
    print(total)
    precision = total['correct_recs']/total['total_recs']
    recall = total['correct_recs']/total['total_correct']
    print(f'precision: {precision}\nrecall: {recall}')
```

```python
calculate_metrics(results, 2)
calculate_metrics(results, 50)
```
