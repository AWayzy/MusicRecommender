```python
import pandas as pd
import numpy as np
import scipy.sparse
```

```python
df = pd.read_csv('500playlists.csv', low_memory=False)
df = df.dropna(axis=1, inplace=False)
df
```

```python
from mlxtend.frequent_patterns import apriori


freqItems = apriori(df, min_support=0.01, use_colnames=True)
```

```python
freqItems
```

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(freqItems, metric="confidence", min_threshold=0.7)
rules
```

```python
df = pd.read_csv('5000playlists.csv', low_memory=False)
df = df.dropna(axis=1, inplace=False)
df
```

```python
from mlxtend.frequent_patterns import fpgrowth
freqItems = fpgrowth(df, min_support=0.002, use_colnames=True)
```

```python
freqItems['length'] = freqItems['itemsets'].apply(lambda x: len(x))
```

```python
freqItems.dropna(axis=0, inplace=True)
```

```python
freqItems
```

```python
rules = association_rules(freqItems, metric="confidence", min_threshold=0.7)
rules
```

```python
rules = rules[rules.consequents != frozenset({'nan'})]
rules = rules[rules.antecedents != frozenset({'nan'})]
rules
```

```python
writeRules = pd.DataFrame(rules['antecedents'].apply(lambda x: list(x)).astype("unicode"))
writeRules['consequents'] = rules['consequents'].apply(lambda x: list(x)).astype("unicode")
```

```python
writeRules
```

```python
writeRules.to_csv('FPGrowth0.7Rules.csv', index = False)
```
